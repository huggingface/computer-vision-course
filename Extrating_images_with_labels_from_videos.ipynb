```python
import cv2
import os

# Paths
video_path = '/content/drive/MyDrive/Colab Notebooks/flood shared with Amir/flood videos/Flood In Booni Upper Chitral Pakistan_part_1.avi'
output_images_dir = '/content/drive/MyDrive/Water Segmentation/frames/train/images'
output_labels_dir = '/content/drive/MyDrive/Water Segmentation/frames/train/labels'

# Create output directories if they don't exist
os.makedirs(output_images_dir, exist_ok=True)  # Create images output directory if it doesn't exist
os.makedirs(output_labels_dir, exist_ok=True)  # Create labels output directory if it doesn't exist

# Open video
cap = cv2.VideoCapture(video_path)
frame_rate = cap.get(cv2.CAP_PROP_FPS)  # Frames per second
frame_count = 0
save_frequency = int(frame_rate)  # Modify to control frame save interval

# Example bounding boxes: (x_min, y_min, x_max, y_max, class_id)
# You need to replace this with actual logic to get bounding boxes if necessary
frame_boxes = {
    0: [(10, 10, 50, 50, 1)],  # frame_number: [(x_min, y_min, x_max, y_max, class), ...]
    1: [(20, 20, 60, 60, 1)],
    # Add more frames with their bounding boxes as needed
}

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Check if we should save this frame
    if frame_count % save_frequency == 0:
        frame_filename = os.path.join(output_images_dir, f'frame_{frame_count}.jpg')

        # Get bounding boxes for the current frame (or set empty if none)
        boxes = frame_boxes.get(frame_count, [])

        # Draw bounding boxes on the frame
        for (x_min, y_min, x_max, y_max, cls) in boxes:
            cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)  # Draw rectangle
            cv2.putText(frame, str(cls), (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)  # Class label

        # Save the labeled frame as an image
        cv2.imwrite(frame_filename, frame)

        # Save bounding boxes for this frame in a text file
        label_filename = os.path.join(output_labels_dir, f'frame_{frame_count}.txt')
        with open(label_filename, 'w') as label_file:
            for (x_min, y_min, x_max, y_max, cls) in boxes:
                label_file.write(f"{cls} {x_min} {y_min} {x_max} {y_max}\n")

    frame_count += 1

cap.release()
```
